```python
import simpy
def main():
    
    env = simpy.Environment()
    env.process(bank_row(env))
    env.run(until = 480)
    print('Simulacion completa')

def bank_row(env):
    while True:
        print("El cajero estara libre en t="  + str(env.now) + ' min')
        yield env.timeout(1)
        print("El cajero esta atendiedo en t= " + str(env.now) +' min')
        yield env.timeout(10)
        
if __name__ ==  '__main__':
    main()
    
```
